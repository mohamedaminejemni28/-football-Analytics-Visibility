```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Arc, Rectangle, Circle

# Load your data (replace with your actual data loading)
# df = pd.read_csv('player_data.csv')
# player1_data = df[df['player_id'] == 1]

# Sample data for demonstration
player1_data = pd.DataFrame({
    'position_x': [575, 580, 590, 600, 610, 620, 630, 640, 650, 660],
    'position_y': [683, 690, 700, 710, 720, 730, 740, 750, 760, 770],
    'frame_num': range(10)
})

# Create styled pitch
def create_styled_pitch():
    fig, ax = plt.subplots(figsize=(14, 9))
    fig.set_facecolor('#f5f5f5')
    ax.set_facecolor('#2e8b57')  # Pitch green
    
    # Set boundaries based on your coordinate system
    ax.set_xlim(0, 1200)
    ax.set_ylim(0, 800)
    ax.axis('off')  # Hide axes
    
    # Pitch outline
    plt.plot([0, 0, 1200, 1200, 0], [0, 800, 800, 0, 0], color='white', linewidth=2)
    
    # Center line
    plt.plot([600, 600], [0, 800], color='white', linewidth=2)
    
    # Center circle
    center_circle = Circle((600, 400), 70, color='white', fill=False, linewidth=2)
    ax.add_patch(center_circle)
    
    # Penalty areas
    plt.plot([0, 180, 180, 0], [300, 300, 500, 500], color='white', linewidth=2)  # Left
    plt.plot([1200, 1020, 1020, 1200], [300, 300, 500, 500], color='white', linewidth=2)  # Right
    
    # Goals
    plt.plot([0, -20, -20, 0], [350, 350, 450, 450], color='white', linewidth=3)  # Left
    plt.plot([1200, 1220, 1220, 1200], [350, 350, 450, 450], color='white', linewidth=3)  # Right
    
    return fig, ax

# Create pitch
fig, ax = create_styled_pitch()

# Plot player path with style
x = player1_data['position_x']
y = player1_data['position_y']
ax.plot(x, y, 'w-', linewidth=3, alpha=0.7, label='Movement Path')  # White path line
ax.scatter(x, y, c=player1_data['frame_num'], cmap='YlOrRd', s=100, 
           edgecolors='black', linewidths=1, label='Position (warmer = later frame)')

# Add frame numbers
for i, row in player1_data.iterrows():
    ax.annotate(str(row['frame_num']), 
                (row['position_x'], row['position_y']),
                color='black', fontsize=8, ha='center', va='center')

# Add stats box like your example
stats_text = "Total Distance: 120m\nAvg Speed: 7.2 km/h\nTop Speed: 9.1 km/h"
ax.text(50, 700, stats_text, color='white', fontsize=12, 
        bbox=dict(facecolor='black', alpha=0.7, edgecolor='white'))

# Add match stats like your example
match_stats = "Total Shots: 20\nxG: 0.85\nGoals: 0"
ax.text(1000, 700, match_stats, color='white', fontsize=12, weight='bold',
        bbox=dict(facecolor='#1a3a6e', alpha=0.9, edgecolor='white'))

# Add legend and title
ax.legend(loc='upper right', facecolor='black', edgecolor='white', 
          labelcolor='white', fontsize=10)
plt.title("Player 1 Movement Analysis", color='white', fontsize=16, pad=20)

# Add club badge (example)
club_logo = plt.imread('club_badge.png')  # Replace with your image
imagebox = OffsetImage(club_logo, zoom=0.15)
ab = AnnotationBbox(imagebox, (1100, 100), frameon=False)
ax.add_artist(ab)

plt.tight_layout()
plt.savefig('player1_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
